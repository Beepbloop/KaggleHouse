```python
import pandas as pd
import os
import seaborn as sns
%pylab inline
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
for dirname, _, filenames in os.walk('../../Datas/KaggleHouse/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
def printNewDF():
    print("================================================================================")
    print("*                                                                              *")
    print("*                                 NEW DATA FRAME                               *")
    print("*                                                                              *")
    print("================================================================================")
```

```python
train = pd.read_csv('../../Datas/KaggleHouse/train.csv')
test = pd.read_csv('../../Datas/KaggleHouse/test.csv')
data_list = [train, test]
```

```python
print("Example Data")
def exampleData(df):
    printNewDF()
    print(df.head(1))
```

```python
for df in data_list:
    exampleData(df)
```

```python
def precNan(df):
    df_na_col = df.columns[df.isna().any()]
    print(df[df_na_col].isna().sum()/df.shape[0]*100)
```

```python
for df in data_list:
    printNewDF()
    precNan(df)
```

```python
def plotAllHis(df):
    for col in df.columns.to_list():
        print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
        try:
            sns.distplot(df[col].dropna(),label=col)
            print(col + "dtype: "+ str(df[col].dtype))
            plt.show()
        except:
            print(col + "Unique: " + str(train[col].unique()))
            print("Num of unique: "+ str(train[col].unique().shape[0]))
```

```python
for df in data_list:
    printNewDF()
    plotAllHis(df)
```

```python
def plotHeat(df):
    plt.figure(figsize=[30,20])
    sns.heatmap(df.select_dtypes(exclude=['object']).dropna().corr(), annot=True)
```

```python
for df in data_list:
    plotHeat(df)
```
